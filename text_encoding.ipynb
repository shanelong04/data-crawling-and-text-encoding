```python
# Text normalization (lowercasing, punctuation removal)
import string

remove_characs = '\t“”' + string.punctuation
def text_normalize(text):
    text = text.lower()
    text = text.strip()
    text = text.replace('\n', ' ')
    for char in remove_characs:
        text = text.replace(char, '')
    
    return text
```

```python
# Create dictionary
def create_dic(corpus):
    dictionary = []
    for para in corpus:
        para = text_normalize(para)
        tokens = set(para.split())
        dictionary.extend(tokens)

    return dictionary
```

```python
# Create new text representation (index-base encoding)
def vecterize(text, dictionary, unknown_token_id):
    text = text_normalize(text)
    tokens = text.split()
    vector = [dictionary.index(token) if token in dictionary else unknown_token_id for token in tokens]

    return vector
```

```python
import os

# create paragraphs list
def create_para_corpus(root_dir):
    para_corpus = []
    for file_name in os.listdir(root_dir):
        filepath = os.path.join(root_dir, file_name)

        with open(filepath, 'r') as f:
            para = f.read()
            para_corpus.append(para)
    
    return para_corpus
```

```python
# Test on vn_news_corpus
root_dir = 'vn_news_corpus'
dictionary = create_dic(create_para_corpus(root_dir))
print(f'Dictionary: {dictionary}\nNumber of tokens: {len(dictionary)}')
```

```python
text = 'Xin chào các bạn tôi tên là Long'
vecterize(text, dictionary, len(dictionary))
```
